```python
class Manusia:
    def __init__(self, warna_rambut, tinggi = 165, berat = 60):
        self.warna_rambut = warna_rambut
        self.tinggi = tinggi
        self.berat = berat
    
    def cat_rambut(self, warna):
        self.warna_rambut = warna
        print(f"Warna Rambut Berubah Menjadi {warna}")
    
    def BMI(self):
        return self.berat / self.tinggi ** 2
```

```python
hekal = Manusia(warna_rambut="Hitam", tinggi=170, berat=75)
```

```python
hekal.berat
```

```python
hekal.cat_rambut("biru")
```

```python
hekal.warna_rambut
```

```python
hekal.BMI()
```

# Inheritance

```python
class Mahasiswa(Manusia):
    def __init__(self, warna_rambut, tinggi, berat, univ):
        super().__init__(warna_rambut, tinggi, berat)
        self.universitas = univ
    
    def pindah_univ(self, univ_baru):
        self.universitas = univ_baru
```

```python
terry = Mahasiswa("hitam", 180, 80, "Poliban")
```

```python
terry.universitas
```

```python
terry.pindah_univ("MIT")
```

```python
terry.universitas
```

```python
class CustomNN(nn.Module):
    def __init__(self, input_size, output_size):
        super().__init__()
        self.fc = nn.Sequential(
            nn.Linear(input_size, 16),
            nn.ReLU(),
            nn.Linear(16, 8),
            nn.ReLU(),
            nn.Linear(8, output_size),
            nn.Sigmoid(),
        )
    
    def forward(self, X):
        return self.fc(X)
```

```python
model = CustomNN(7, 1)
```

```python
contoh_data = torch.rand(100, 7)
```

```python
output = model.forward(contoh_data)
output
```
